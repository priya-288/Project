```python
import pandas as pd
```

```python
df=pd.read_csv("FashionDataset.csv")
```

```python
## Removing Duplicates
df=df.drop_duplicates()
## Rows after removing duplicate
df.shape
```

```python
## Checking the datatype of the column
df.dtypes
```

```python
##Checking the Null values(But data has 'NaN' as string)
df.isnull().sum()
```

```python
##Striping of the unessary part
df["MRP"]=df["MRP"].str.strip("Rs\n")
```

```python
df.head(5)
```

```python
## changing the data type 
df['MRP']=pd.to_numeric(df['MRP'],errors='coerce')
```

```python
## checking the null value in MRP column
df['MRP'].isnull().sum()
```

```python
## Creating the new Column and replacing null
df['MRP_'] = df['MRP'].fillna(df.groupby('Category')['MRP'].transform('mean'))
```

```python
##Droping the mrp column
df.drop('MRP',axis=1,inplace=True)
```

```python
## discount
df['Discount']=df['Discount'].str.strip("% off")
```

```python
df['Discount']=pd.to_numeric(df['Discount'],errors='coerce')
```

```python
df['Discount_percent']=df['Discount'].fillna(
    df.groupby('BrandName')['Discount'].transform('median')
    ).fillna(0)
```

```python
df.drop('Discount',axis=1,inplace=True)
```

```python
## SellPrice
df['SellPrice']=pd.to_numeric(df['SellPrice'],errors='coerce')
```

```python
df['SellPrice_']=df['SellPrice'].fillna(df['MRP_']*(df['Discount_percent']/100))
```

```python
df.drop('SellPrice',axis=1,inplace=True)
```

```python
## Cheaking the Category Column
df['Category'].unique()
```

```python
## Replacing the unessary part and renaming the column
df['Category']=df['Category'].str.replace('-Women','')
```

```python
df.rename(columns={'Category':'Womens_Category'},inplace=True)
```

```python
## Droping the UnNamed column
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df['Sizes'].unique()
```

```python
## Replacing the string Nan NaN and fill with pd.NA(actual null)
df['Sizes']=df['Sizes'].replace(['nan','NAN','Nan',''],pd.NA)
```

```python
## Filling the na with free size as as the data of null is too large we can not delete
df['Sizes']=df['Sizes'].fillna('Free Size')
```

```python
## checking the data
df.head(10)
```

```python
df=df.replace(['Nan','nan',''],pd.NA)
```

```python
df=df.dropna()
```

```python
df = df.reset_index(drop=True)
```
